```python
import math
import csv
import numpy as np
import pandas as pd 

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
get_ipython().magic('matplotlib inline')
```

```python
#1440 Min Bars - Daily
df = pd.read_csv("/Users/mikayla/Desktop/ESRaw.txt", index_col = 0, parse_dates=True)
```

```python
df = df['Close']
```

```python
df = pd.DataFrame(df)
```

```python
df.head()
```

```python
df['n day'] = df['Close'].diff(40).abs()
df['1Day Diff'] = df['Close'].diff(1).abs()
df['Sum Diff'] = df['1Day Diff'].rolling(40).sum()
df['EFF Ratio'] = df['n day']/df['Sum Diff']
df.dropna(inplace=True)
df.tail()
```

```python
df['EFF Ratio'] .plot(figsize=(20,10), legend=True, linestyle='--', marker='o')
```
